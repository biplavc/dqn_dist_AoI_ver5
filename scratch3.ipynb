```python
import pandas as pd

path = "/home/biplav/AoI/old_dist_AoI_ver4/models/NPL_text_files/3_text_files/"
fairness = []

def get_fairness(path_new):
    data = pd.read_csv(path_new + '/dict.txt', sep = '\t', header = None)
    x = np.array(data.iloc[:,1])
    # data.info() ## .iloc[:, 1]
    numerator = (np.sum(x))**2
    denominator = len(x) * np.sum([y**2 for y in x])
    return numerator/denominator


def get_path(sch):
    if sch == "MAD":
        path_new = path + "MAD"
        fairness_new = get_fairness(path_new)
    elif sch == "MAF":
        path_new = path + "greedy"
        fairness_new = get_fairness(path_new)    
    elif sch == "RR":
        path_new = path + "round_robin"
        fairness_new = get_fairness(path_new) 
    elif sch == "PF":
        path_new = path + "proportional_fairness"
        fairness_new = get_fairness(path_new) 
    elif sch == "random":
        path_new = path + "random"
        fairness_new = get_fairness(path_new) 

    return fairness_new

fairness.append(get_path("MAD"))
fairness.append(get_path("MAF"))
fairness.append(get_path("RR"))
fairness.append(get_path("PF"))
fairness.append(get_path("random"))

print(fairness)
```
